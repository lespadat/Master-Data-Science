```python
# siempre empezamos importando las librerias con las q vamos a trabajar

import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importar el csv (OJO con el directorio del csv)
df=pd.read_csv("vtaTotTienda2.csv",sep=";")
```

```python
df.head(10)
```

```python
df["pesoPromo"]=df["vtapromo"]/df["vtatot"]
```

```python
df["pesoPromo"]=df["pesoPromo"]*100
```

```python
df3U=df[df["tienda"]=='003U']
```

```python
df3U=df3U.sort("nmes",ascending=True)
```

```python
plt.plot(df3U["pesoPromo"])
plt.ylim(0,(df3U["pesoPromo"]).max()+3)       # escala del eje y de 0 al valor máx de la serie+3000
plt.show()
```
